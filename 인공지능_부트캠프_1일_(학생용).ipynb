##### 튜플에 포함된 아이템은 수정할 수 없다.
##### 선언 방법: [  ] 대신 (  )
##### 삽입, 삭제, 정렬 등의 기능은 없고 조회 기능만 주로 사용한다.

````python
fruits=('사과','포도','바나나','자두','배')

print(fruits)
print(type(fruits))
````

````python
fruits=('사과','포도','바나나','자두','배')
odd_item=[]
for i in range(len(fruits)):
    if i %2 ==1:
        odd_item.append(fruits[i])
print(odd_item) #리스트
print(fruits[1::2]) #튜플
````

````python
color=('red','yellow','orange','pink','blue','black')
print('color의 자료형',type(color))

color=list(color)
print('color의 자료형',type(color))

color=tuple(color)
print('color의 자료형',type(color))
````

# 클래스와 객체
#### 객체 지향 프로그래밍(OOP)이란?
###### 객체를 이용하여 프로그램을 만드는 것
###### - 객체는 속성과 기능으로 구성되어있다.
###### - 속성은 객체를 '구성하는 요소', 기능은 '객체가 하는 행위'를 의미한다.
###### - ex) 계산기=객채, 속성: 숫자버튼, 화면, 기능: 더하기, 빼기, 곱하기, 나누기

***
###### - 클래스 = 설계도 / 객체 = 설계도로 만든 제품 / 속성 = 클래스안의 변수
###### - 메서드 = 클래스안의 함수 / 생성자 = 객체를 만들 때 실행되는 함수
###### - 생성자 = 객체를 만들 때 실행되는 함수


***
###### 객체를 생성하기 위해서는 클래스가 필요하다.
###### - 클래스란 객체를 생성하기 위한 틀
###### - 클래스의 기능을 표현하는 사용자 정의 함수를 메서드 라고 한다.
###### ex) 캐릭터 클래스에서 모든 캐릭터는 체력과 공격력이라는 속성을 가져야 한다.
######     클래스를 바탕으로 만든 검술사, 요정, 도적이라는 개체는 속성에 구체적이고 개별적인 값을 가진다.
######     검술사의 체력은 100, 공격력은 10 / 요정은 각각 60, 5 / 도적은 85, 8

## **생성자와 인스턴스 속성**
###### 속성은 객체가 공통으로 가지는 특성
###### 속성에는 클래스 속성, 인스턴스 속성이 있는데 인스턴스 속성을 많이 사용
###### 검술사, 요정, 도적은 모두 체력과 공격력이라는 속성을 가지고 있음
###### 캐릭터별로 체력과 공격력이 다르다.
###### 객체별로 개별적인 값을 가진 속성을 ***인스턴스 속성***이라고 한다.
###### 인스턴스 속성은 객체명.속성으로 접근할 수 있음


***
###### 체력의 속성명을 strength, 공격력의 속성명을 attack_damage라 할 때
###### 검술사의 체력은 swordman.strength, 요정의 공격력은 fairy.attak_damage
***


###### 이렇게 객체별로 속성에 개별적인 값을 저장하려면 생성자라는 특별한 메서드를 정의해야 함.
###### __init__()이 바로 생성자!




```
def __init__(): 클래스 내부에서 사용하면 해당 메서드가 생성자가 된다.
__init__() 역시 메서드이므로 self를 첫 번째 매개변수로 넣어야 한다.
```



````python
class Character:
    def __init__(self, name, hp, ad):
        self.name=name
        self.strength=hp
        self.attack_damage=ad
        print('Character 클래스의 객체 생성 ->', self.name)
    def attack(self, target):
        print(target, '을/를 공격합니다.')

swordsman=Character('검술사', 100, 10)
print(swordsman.name, swordsman.strength, swordsman.attack_damage)
````

````python
class Character:
    def __init__(self, name, hp, ad):
        self.name=name
        self.strength=hp
        self.attack_damage=ad
        print('Character 클래스의 객체 생성 ->', self.name)
    def attack(self, target):
        print(target, '을/를 공격합니다.')
    def information(self):
        print('이름 ->',self.name, ', 체력 ->', self.strength, ', 공격력 ->', self.attack_damage)

swordsman=Character('검술사', 100, 10)

print(swordsman.name, swordsman.strength, swordsman.attack_damage)
swordsman.information() #print()처럼 하기 귀찮으니 information을 활용하자

print('='*50)

fairy=Character('요정', 60, 40)
fairy.information()
thief=Character('도적', 70, 8)
thief.information()
````

````python
# 메서드를 추가하여 인스턴스 속성 변경하기

class Character:
    def __init__(self, name, hp, ad):
        self.name=name
        self.strength=hp
        self.attack_damage=ad
        print('Character 클래스의 객체 생성 ->', self.name)
    def attack(self, target):
        print(target, '을/를 공격합니다.')
    def information(self):
        print('이름 ->',self.name, ', 체력 ->', self.strength, ', 공격력 ->', self.attack_damage)

swordsman=Character('검술사', 100, 10)
print(' 변경 전 ----------------')
swordsman.information()

swordsman.strength=120
swordsman.attack_damage=20

print(' 변경 후 ----------------')
swordsman.information()
````

###tkinter


Tkinter는 그래픽 사용자 인터페이스(GUI)를 만들기 위한 표준 Python 라이브러리입니다. 그래픽 인터페이스를 만들기 위한 도구와 위젯의 집합인 Tk 툴킷을 기반으로 하며, Python 애플리케이션을 위한 창, 대화 상자, 버튼, 레이블 및 기타 위젯을 간단하게 만들 수 있는 방법을 제공합니다.

````python
import tkinter as tk

# 기본 윈도우 생성
root = tk.Tk()
root.title("Simple GUI")

# 라벨 생성
label = tk.Label(root, text="Hello, Tkinter!")
label.pack()

# 버튼 생성
button = tk.Button(root, text="Click Me", command=lambda: label.config(text="Button Clicked!"))
button.pack()

# 슬라이더(Scale) 생성
slider = tk.Scale(root, from_=0, to=100, orient="horizontal")
slider.pack()

# 메인 루프 실행
root.mainloop()
````

````python

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
````

````python
# 사인파를 그리는 함수 정의
def plot_sine_wave(frequency, amplitude):
    x = np.linspace(0, 2 * np.pi, 1000)
    y = amplitude * np.sin(frequency * x)

    plt.figure(figsize=(10, 4))
    plt.plot(x, y)
    plt.ylim(-10, 10)
    plt.title(f"Sine Wave: Frequency={frequency}, Amplitude={amplitude}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()

# 주파수 슬라이더
frequency_slider = widgets.FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='Frequency:')
# 진폭 슬라이더
amplitude_slider = widgets.FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='Amplitude:')

# 인터랙티브 위젯을 사용해 그래프 업데이트
widgets.interactive(plot_sine_wave, frequency=frequency_slider, amplitude=amplitude_slider)
````

### 일반화 시키기

````python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

class WavePlotter:
    def __init__(self, frequency=1.0, amplitude=1.0):
        self.frequency = frequency
        self.amplitude = amplitude

        # 주파수와 진폭 슬라이더 생성
        self.frequency_slider = widgets.FloatSlider(value=self.frequency, min=0.1, max=10.0, step=0.1, description='Frequency:')
        self.amplitude_slider = widgets.FloatSlider(value=self.amplitude, min=0.1, max=10.0, step=0.1, description='Amplitude:')

        # 슬라이더 값이 변경될 때 plot을 업데이트
        self.frequency_slider.observe(self.update_plot, names='value')
        self.amplitude_slider.observe(self.update_plot, names='value')

        # 초기 그래프 설정
        self.fig, self.ax = plt.subplots(figsize=(10, 4))
        self.line, = self.ax.plot([], [])
        self.ax.set_ylim(-10, 10)
        self.ax.set_xlabel("x")
        self.ax.set_ylabel("y")
        self.ax.grid(True)

        # 초기 플롯 업데이트
        self.update_plot()

    def wave_function(self, x):
        """기본적으로 이 함수는 오버라이드 되어야 합니다."""
        raise NotImplementedError("wave_function method must be overridden in subclasses.")

    def update_plot(self, change=None):
        self.frequency = self.frequency_slider.value
        self.amplitude = self.amplitude_slider.value

        x = np.linspace(0, 2 * np.pi, 1000)
        y = self.amplitude * self.wave_function(self.frequency * x)

        self.line.set_data(x, y)
        self.ax.set_xlim(0, 2 * np.pi)
        self.ax.set_ylim(-10, 10)
        self.ax.figure.canvas.draw()

    def display(self):
        display(self.frequency_slider, self.amplitude_slider)
        display(self.fig)
````

````python
class SinWavePlotter(WavePlotter):
    def wave_function(self, x):
        return np.sin(x)

    def __init__(self, frequency=1.0, amplitude=1.0):
        super().__init__(frequency, amplitude)
        self.ax.set_title("Sine Wave")
````

````python
class CosWavePlotter(WavePlotter):
    def wave_function(self, x):
        return np.cos(x)

    def __init__(self, frequency=1.0, amplitude=1.0):
        super().__init__(frequency, amplitude)
        self.ax.set_title("Cosine Wave")
````

````python
class TriangleWavePlotter(WavePlotter):
    def wave_function(self, x):
        return 2 * np.arcsin(np.sin(x)) / np.pi

    def __init__(self, frequency=1.0, amplitude=1.0):
        super().__init__(frequency, amplitude)
        self.ax.set_title("Triangle Wave")
````

````python
# 사인파 플로터 인스턴스
sin_plotter = SinWavePlotter()
sin_plotter.display()

# 코사인파 플로터 인스턴스
cos_plotter = CosWavePlotter()
cos_plotter.display()
````
