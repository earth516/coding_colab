### 컴퓨터는 어떻게 사람의 말을 이해할까?

image.png

```python
# 필요한 라이브러리 불러오기
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# 간단한 텍스트 데이터
texts = [
    "I love this movie",
    "This is a great film",
    "Amazing movie, really enjoyed it",
    "I did not like this movie",
    "The film was boring",
    "I hated this movie",
    "This movie is fantastic",
    "Terrible movie, waste of time",
    "Best film I have ever seen",
    "Not a good movie"
]

# 라벨 (1 = 긍정적, 0 = 부정적)
labels = [1, 1, 1, 0, 0, 0, 1, 0, 1, 0]

# CountVectorizer를 사용해 텍스트를 숫자로 변환
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(texts)

# 데이터를 학습용과 테스트용으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, labels, test_size=0.3, random_state=42)

# 나이브 베이즈 분류 모델 생성 및 학습
model = MultinomialNB()
model.fit(x_train, y_train)

# 모델로 예측
y_pred = model.predict(x_test)

# 정확도 출력
print("테스트 정확도:", accuracy_score(y_test, y_pred))
```


테스트 정확도가 0.3333...이라는 것은 모델이 테스트 데이터에서 약 33.33%의 정확도로 예측을 했다는 의미입니다. 이는 테스트 데이터에 대해 모델이 얼마나 잘 예측했는지를 나타내는 지표입니다.

1. 정확도의 의미
정확도(accuracy)는 모델이 올바르게 예측한 데이터 포인트의 비율을 의미합니다. 즉, 총 테스트 데이터 중에서 모델이 얼마나 많은 데이터를 올바르게 예측했는지를 퍼센트로 나타낸 값입니다.

2. 정확도가 낮다는 것은?
테스트 정확도가 0.3333...라는 것은 모델이 테스트 데이터의 약 33.33%만을 정확하게 분류했다는 의미입니다. 이는 모델이 부정확하게 작동하고 있음을 나타냅니다. 이유는 여러 가지가 있을 수 있습니다:

데이터가 적음: 예제에서 사용된 데이터셋은 매우 작고, 문장도 단순합니다. 학습에 사용된 데이터가 적으면 모델이 충분히 학습되지 않아 새로운 데이터를 잘 예측하지 못할 수 있습니다.
모델의 단순함: 간단한 나이브 베이즈 모델이 텍스트의 복잡한 의미를 잘 잡아내지 못했을 수 있습니다. 더 복잡한 모델이나 다른 접근 방식을 사용하면 정확도가 높아질 수 있습니다.
데이터 분리의 특성: 데이터가 학습용과 테스트용으로 나눠졌을 때, 학습 데이터가 적절하게 대표성을 가지지 못했다면, 모델이 학습에 사용되지 않은 새로운 데이터에 대해 잘 예측하지 못할 수 있습니다.

4. 결론
테스트 정확도가 0.3333...이라는 것은 모델이 테스트 데이터에서 제대로 된 예측을 하지 못하고 있다는 신호입니다. 정확도를 높이기 위해서는 더 많은 데이터를 사용하거나, 모델을 개선하거나, 데이터 전처리를 개선하는 방법 등을 고려할 수 있습니다.

```python
# 새로운 문장을 예측하기 위해 추가적인 문장 준비
new_texts = [
    "I really enjoyed the movie",
    "It was a terrible experience",
    "The plot was boring and predictable",
    "What a fantastic performance!",
]

# 앞서 사용한 vectorizer를 통해 새 문장도 동일한 방식으로 변환
new_x = vectorizer.transform(new_texts)

# 학습된 모델을 사용하여 예측
new_predictions = model.predict(new_x)

# 예측 결과 출력
for text, prediction in zip(new_texts, new_predictions):
    sentiment = "긍정적" if prediction == 1 else "부정적"
    print(f"문장: '{text}' -> 예측 결과: {sentiment}")
```

```python
# 새로운 문장을 예측하기 위해 추가적인 문장 준비
new_texts = [
    "I really enjoyed the movie",   # 예상: 긍정적
    "It was a terrible experience", # 예상: 부정적
    "The plot was boring and predictable", # 예상: 부정적
    "What a fantastic performance!", # 예상: 긍정적
    "The movie was not bad, but it could have been better", # 애매한 문장
]

# 앞서 사용한 vectorizer를 통해 새 문장도 동일한 방식으로 변환
new_x = vectorizer.transform(new_texts)

# 학습된 모델을 사용하여 예측
new_predictions = model.predict(new_x)

# 예측 결과 출력
for text, prediction in zip(new_texts, new_predictions):
    sentiment = "긍정적" if prediction == 1 else "부정적"
    print(f"문장: '{text}' -> 예측 결과: {sentiment}")
```
